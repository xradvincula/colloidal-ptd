```python
#Load CVs corresponding to the initial crystallising trajectory (reference)
cvs_13 = np.loadtxt(loading_path)
```

```python
plt.xlabel('$n$');
plt.ylabel('$n(q6)$');
cmap='coolwarm',
sc=plt.scatter(cvs_13[:,1], cvs_13[:,3], c=cvs_13[:,0], cmap="coolwarm", s=40, marker=".")
plt.xlim(0.0,)
plt.ylim(0.0,)
plt.colorbar(sc,label='Simulation time [$\\tau^*$]')
plt.savefig(storing_path +"trajectory_cv_1_1.png", dpi=700, bbox_inches='tight')
```

```python
plt.xlabel('$\widetilde{n}$');
plt.ylabel('$\widetilde{n}(q6)$');
sc=plt.scatter(np.power(cvs_13[:,1], 1/3), np.power(cvs_13[:,3], 1/3), c=cvs_13[:,0], cmap="coolwarm", s=40, marker=".")
plt.xlim(0, 7);
plt.ylim(0, 3);
plt.colorbar(sc,label='Simulation time [$\\tau^*$]')
plt.savefig(storing_path +"trajectory_cv_1_2.png", dpi=700, bbox_inches='tight')
```
